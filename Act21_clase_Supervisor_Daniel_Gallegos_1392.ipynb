```python
class Supervisor:
    def __init__(self, nombre, telefono, edad):
        # Atributos de la clase
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad

    # Función 1: Organizar tareas
    def organizar_tareas(self):
        return f"{self.nombre} está asignando las tareas del día al equipo."

    # Función 2: Evaluar desempeño
    def evaluar_desempeno(self):
        return f"El supervisor {self.nombre} está revisando las métricas de rendimiento."

    # Función 3: Capacitar
    def capacitar(self):
        return f"{self.nombre} ha iniciado una sesión de capacitación para los nuevos empleados."

# --- Creación de Objetos (Instancias) ---

# Creación del Supervisor 1 con datos de ejemplo
supervisor1 = Supervisor("Laura García", "555-0123", 35)

# Creación del Supervisor 2 con datos de ejemplo
supervisor2 = Supervisor("Carlos Ruiz", "555-9876", 42)

# --- Ejecución y Muestra de Resultados ---

print(f"--- Ficha del Supervisor 1 ---")
print(f"Nombre: {supervisor1.nombre}")
print(f"Teléfono: {supervisor1.telefono}")
print(f"Edad: {supervisor1.edad}")
print(f"Acción 1: {supervisor1.organizar_tareas()}")
print(f"Acción 2: {supervisor1.evaluar_desempeno()}")
print(f"Acción 3: {supervisor1.capacitar()}")

print("\n" + "-"*30 + "\n")

print(f"--- Ficha del Supervisor 2 ---")
print(f"Nombre: {supervisor2.nombre}")
print(f"Teléfono: {supervisor2.telefono}")
print(f"Edad: {supervisor2.edad}")
print(f"Acción 1: {supervisor2.organizar_tareas()}")
print(f"Acción 2: {supervisor2.evaluar_desempeno()}")
print(f"Acción 3: {supervisor2.capacitar()}")

print("Programa creado por Daniel Gallegos NL = 20")
```
